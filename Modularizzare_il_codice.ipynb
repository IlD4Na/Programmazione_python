# Introduzione ai moduli


Modulo : E' un file che contiene definizioni (funzione,classi,etc...) che possono essere utilizzate all'interno del nostro programma, senza doverle ri-definire.

Cerchiamo di ricordare la classe Square che abbiamo creato:



```
class Square:

   def __init__(self,l):
       self.l = l

   def area(self,l):
       return self.l**2

   def perimeter(self,l):
       return self.l*4
```

Immaginiamo di prendere questa classe e memorizzarla in un file chiamato square.py (file di estensione .py).

Abbiamo il nostro script principale che si chiamera main.py che useremo per avviare il nostro programma e all'interno di questo scriveremo:



```
import square

sqr = square.Square(4)
```

Usando la keyword import possiamo importare il modulo. Il nome del modulo  il nome che abbiamo dato al file senza l'estensione .py.

per utilizzare dobbiamo fare:

<mark> nome_del_modulo.nome_della_funzione_della_classe

Quando all'interno di una cartella (package) andiamo a creare un modulo intitolato init.py stiamo dicendo a python di utilizzare quell'init per convogliare tutti gli altri moduli.


```
def sqrt(n):

  """
  
  Calcolo della radice quadrata di un numero

  Positional arguments:

  n : float  -- il numero di cui calcolare al radice quadrata

  
  """
  return n**(1/2)

def pow(n, p=2):

  """
  Calcolo di una potenza di un numero

  Positional arguments:
  n : float -- il numero di cui calcolare la potenza

  Keyword argument:
  p : float (default 2) -- la potenza da calcolare

  """
  return n**p

def abs(n):
  """
  Calcolo del valore assoluto

  Positional arguments
  n : float -- il numero di cui calcolare il valore assoluto

  """
  return n if n>=0 else -n



def factorial(n):

  
  fact = 0
  
  for i in range(1, n+1):
    fact*=i #fact = fact * i
  return fact
```

Qui si  parlato di un errore di import circolare, nel caso in cui nel modulo __init__.py noi importiamo distance e nel modulo distance.py noi importiamo funzioni di __init__.py

Cosa  stato fatto ?

All'interno del file __init__.py si mettono tutti gli import tipo



```
from . import distance

from . import arithmetic
```

Quindi avremo 3 moduli:

-__init__.py

-arithmetic.py

-distance.py

quindi nel modulo distance dovremo solo riportare:



```
from arithmetic import sqrt, pow, abs
```
(E' molto importante capire questa cosa)

ovviamente in __init__.py possiamo anche importare tutto in questa maniera in maniera tale che solo importanto my_math possiamo usare tutte le funzioni:



```
# Questo  ==> init.py

from distance import *
from arithmetic import *
```

"*" l'asterisco in informatica quasi semrpe sta a significare che ci vogliamo prendere tutto.

In questa maniera  come se il modulo init fosse un calderone all'interno del quale mettiamo tutto a livello di import.

In questa maniera se vogliamo usare qualcosa potremmo semplicemente usare il nome del package + il nome del metodo/funzione.

Quindi dal package my_math potremmo utilizzare tutte le funzioni dei vari moduli



