# Casting : Portare il contenuto di una variabile da un tipo ad un altro



Prima però vediamo i diversi approcci di vari linguaggi di programmazione.

Abbiamo linguaggi tipizzati, ovvero dove il tipo di una variabile va specificato e una volta definita la variabile potrà contenere solo quel tipo.

**Linguaggio C:**

```
int age = 30;
float height = 177.7;
char gender = 'M';
```
Nella maggior parte dei linguaggi di programmazione il tipo di variabile va dichiarato subito prima del nome della variabile.

**Python** è invece non tipizzato (dinamicamente tipizzato):



```
weight = 80 #int
weight = 80.5 #float
weight = "80.5 Kg" #str
```

Quando il linguaggio non è tipizzato il codice è più flessibile però si rischia facilmente di fare disastri. Anche se si lavora con linguaggi di programmazione non tipizzati è buona abitudine sia a livello documentale che personale specificare dopo il tipo di variabile come commento.


C'è anche la differenza tra tipizzazione debole e tipizzazione forte. In caso di tipizzazione debole il programma potrebbe convertire il tipo di una variabile in un altro senza dircelo.

Un linguaggio debolmente tipizzato è ad esempio il JavaScript.
In un linguaggio fortemente tipizzato il dato non viene convertito in automatico.
Esempio:



```
weight_int = 80 #int
weight_str = "80.5 Kg" #str
weight_int + weight_str # TypeError
```

Ci darà errore perchè non possiamo unire un intero ad una stringa. Ecco che qui interviene il "Casting" che ci permetterà di passare da un tipo di dato all'altro.

```
weight = 80
weight_float = float(weight) # float 80.0
weight_int = int(weight_float) # 80 int
```

# La formattazione


Ci permette di inserire all'interno di una stringa numeri, o anche altre cose con un formato deciso da noi. Ci sono 3 metodiche di formattazione, la prima consiste nell'uso del carattere di formattazione "%".



````python
#USO DEL CARATTERE DI FORMATTAZIONE % SI RIMPIAZZA IL %d CON IL VALORE DELLA VARIABILE age
# %d funge da placeholder

age = 25 #int

print("Hai %d anni" % age)

my_name = input("Come ti chiami? ")
my_age = input("Quanti anni hai? ")
my_age = int(my_age)
my_born = 2024-my_age

print("Il tuo nome è %s, hai %d di anni , quindi sei nato nel %d" % (my_name, my_age, my_born))
````

Ci sono diversi placeholder:


*   %d => per numeri interi
*   %f => per numero con la virgola
*   %s => stringa

Importante sapere che ***bisogna immagazinare le variabili all'interno di una tupla quindi tra*** ().
Quindi questa è la prima via per la formattazione.





````python
my_height = float(input("Inserisci la tua altezza in metri: "))
my_weight = float(input("Inserisci il tuo peso in kg: "))
````

````python
print("Altezza=%.2f m Peso=%.1f kg" % (my_height, my_weight))
````

I float vengono inseriti con ben 6 cifre dopo la virgola. Possiamo però decidere noi il numero di cifre da mostrare. Qui però il float non viene troncato ma arrotondato, non avviene ciò che succede anche nel float.

Vediamo la seconda tecnica che per la formattazione, e anche come specificare le cifre significative in caso le volessimo decidere.

````python
print("Il tuo nome è {name}, hai {age}, quindi sei nato nel {born}"
      .format(name=my_name, age=my_age, born=my_born))
````

````python
print("Altezza={height:.1f} Peso={weight:.0f}".format(height=my_height, weight=my_weight))
````

Invece vediamo adesso la terza metodica per la formattazione. L'ultima tipologia di formattazione è la più semplice e si chiama fStrings.


````python
print(f"Il tuo nome è {my_name}, hai {my_age} anni, quindi sei nato nel {my_born}")

print(f"Altezza={my_height:.1f} Peso={my_weight:.0f}")
````

Queste sono 3 tecniche che si possono usare per formattare su Python, la più usata sono le f strings. Invece quando si usa l'operatore formattazione si va ad usare il metodo classico presente anche in altri linguaggi di programmazione.

Sopra abbiamo visto come passare da una tupla di tuple ad un dizionario con le chiavi e i valori. Vediamo invece come creare un dizionario partendo però da due liste.

````python
names = ["Giuseppe", "Antonio", "Nicola"]
ages = [30, 24, 50]

names_age = zip(names, ages)
names_age

#zip viene usata per combinare più iterabili in un singolo iterabile di tuple

#In questo caso ha preso l'indez 0 di names e l'index 0 di ages e lo ha unito in una tupla

````

<mark>Utilità di zip()
- Combinare dati correlati: È utile quando hai due o più liste che rappresentano dati correlati e vuoi combinarli in coppie o gruppi.

- Iterazione parallela: Facilita l'iterazione parallela su più iterabili.
- Creazione di dizionari: Può essere usato per creare dizionari associando chiavi e valori da due liste separate.

La funzione zip combina le due liste names e ages in una serie di tuple, dove ciascuna tupla contiene un elemento da ciascuna lista. Il risultato è un oggetto di tipo zip, che è un generatore.

Con names e ages come input, zip produce:  <mark>[("Giuseppe", 30), ("Antonio", 24), ("Nicola", 50)].

Python ci permette di usare la funzione zip che ci ritorna un particolare tipo di oggetto, questo viene chiamato "generatore" sono un tipo di oggetto che python utilizza per ottenere dei valori solo quando questi valori effettivamente servono.

<mark>Per passare da quel generatore alla tupla vera e propria basta fare il casting.

````python
names_age_tuple = tuple(names_age) #l'oggetto zip viene convertito in una tupla di tuple

names_age_dict = dict(names_age_tuple) #qui la tupla di tuple viene convertita in un dizionario

print(names_age_dict)
````

IMPORTANTISSIMO PARLARE DI COPIA VS RIFERIMENTO

````python
l1 = ["cane", "gatto", "topo"]
l2 = l1                        #Sia l1 che l2 puntano allo stesso spazio di memoria
l2[1] = "tigre"

l2

````

````python
l1
````

Anche l1 è stata modificata, ma per quale motivo ? Quando noi andiamo a creare un nuovo oggetto quello che fa python è creare uno spazio di memoria in cui andare a salvare quest einformazioni. Poi passa il riferimento dello spazio di memoria alla variabile l1.

Quando invece associamo il valore di una variabile ad un'altra variabile quello che accade è che quello che la seconda variabile conterrà è lo STESSO RIFERIMENTO ALLO SPAZIO DI MEMORIA DOVE SONO STATI SALVATI I DATI DELLA PRIMA MEMORIA.

In sostanza sia l1 che l2 puntano allo stesso oggetto.

SONO QUINDI LO STESSO OGGETTO

Possiamo modificare questa cosa usando la funzione id( )

````python
print(id(l1)) # ci ritorna un codice identificativo dell'oggetto

print(id(l2))

l1 is l2
````

Come possiamo vedere puntano allo stesso oggetto.

Invece se avessimo assegnato a l2 una nuova lista uguale identica senza fare riferimento avremmo avuto due oggetti diversi con id diverso, quindi due spazi di memoria diversi con dentro l'identica informazione. Oppure possiamo usare un metodo di python che ci permette di creare una copia.

In questa maniera non passeremo un riferimento ma creeremo una copia di s1 in s2 con due spazi di memoria diversi.

````python
s1 = ["Sabato", "Domenica", "Mercoledi"]
s2 = s1.copy()

print(id(s1), s1)
print(id(s2), s2)
````

## <mark> __I cicli__ </mark>


__Cicli__ => ci permettono di eseguire delle istruzioni in maniera ciclica

**Ciclo While**


```
vals = [12, 9, 4, 21, 11]
vals_sum = 0

i = 0 # è il contatore

while i < len(vals):
     vals_sum = vals_sum + vals[i]
     i+=1 #significa incremento

vals_mean = vals_sum / len(vals)
```
Questa è una iterazione che continuerà ad essere eseguita finchè la prima espressione di while sarà verificata.

Vediamo come sarebbe stata la soluzione in caso di un ciclo __For__



```
vals = [12, 9, 4, 21, 11]
vals_sum = 0

for val in vals:
    vals_sum = vals_sum + val

vals_mean = vals_sum / len(vals)
```

Suggerimento: tramite il metodo .split() possiamo dividere gli elementi di una stringa in base agli spazi.

Possiamo applicare una trasformazione a tutti gli elementi di una collezione utilizzando la funzione map, questa funzione ritorna un generatore che possiamo riconvertire in una collezione

````python
points = (input("Inserisci i punteggi dei partecipanti divisi da uno spazio: "))

print(f"I punteggi della Giornata dello Sport Universitaria sono i seguenti: {points} ")

points_list = points.split()

l = list(map(int,points_list))

#map ritorna un generatore quindi bisogna sempre fare il casting per riconvertirlo

l.sort()

#sort va a ordinare la lista dal minore al maggiore

print(l[-1])


````

````python
points = (input("Inserisci i punteggi dei partecipanti divisi da uno spazio: "))

print(f"I punteggi della Giornata dello Sport Universitaria sono i seguenti: {points} ")

points_list = points.split()

l = list(map(int,points_list))

for i in l:
  larger = 0
  for j in l:
    if i < j:
      larger+=1

  if larger ==1:
    break

print(i)
````

<mark> 3° ESERCIZIO

Un programma in cui si deve inserire un numero intero che starà a indicare quanti record di studenti vogliamo inserire, questi record saranno caratterizzati dal nome + il voto di 3 materie quali Matematica, Fisica e Chimica. Alla fine possiamo inserire il nome di uno degli studenti e ottenere come output la media.

````python
students ={}

n = int(input("Quanti record di studenti vuoi aggiungere ? "))

a = 0
while a < n:
  student_raw = input("Inserisci il nome dello studente e i voti di Matematica, Fisica e Chimica: ")
  student_added = student_raw.split()
  students[student_added[0]] =[student_added[1],student_added[2],student_added[3]]

  a+=1


call = input("Di quale studente vuoi conoscere la media? ")

mean_student_called = list(map(int, students[call]))

sum = 0

for i in mean_student_called:
  sum+=i

print(sum/len(mean_student_called))

#Oppure possiamo usare una funzione sum(mean_student_called)/len(mean_student_called)

````

<mark>4° Esercizio

Un palindromo è una parola che può essere letta allo stesso modo in entrambi i sensi, sia da sinistra verso destra che da destra verso sinistra.

Tipo: radar, osso, aerea

````python
stop = 0
count = 0
while stop == 0:
  word = input("Inserisci la parola: ")
  if word == "stop":
    stop+=1

  word_actual = word[:]
  word_reverse = word[::-1]

  if word_actual == word_reverse:
    print(f"{word} è una parola palindroma")
  else:
    print(f"{word} non è una parola palindroma")

  count+=1

print(f"Hai verificato che {count} parole fossero palindrome")




````
